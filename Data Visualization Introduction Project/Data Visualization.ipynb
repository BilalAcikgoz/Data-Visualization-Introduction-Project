# SEABORN KÜTÜPHANESİ

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
veri = pd.read_csv("olimpiyatlar_temizlenmiş.csv")
veri.head()
```

# SAÇILIM GRAFİĞİ

```python
def sacilimGrafikCiz():
    sns.scatterplot(x="boy", y="ağırlık", data=veri)
    plt.title("Boy ve Ağırlık İlişkisi - Beyaz Izgara Tema")
    plt.show()
    
sacilimGrafikCiz()    
```

```python
sns.set_style("darkgrid")
sacilimGrafikCiz()
```

```python
sns.set_style("whitegrid")
sacilimGrafikCiz()
```

# Saçılım Grafiği - Kategorik

```python
sns.scatterplot(x="boy", y="ağırlık", hue="madalya", data=veri)
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="ağırlık", hue="cinsiyet", data=veri)
plt.xlabel("Sporcuların Boyu")
plt.ylabel("Sporcuların Ağırlığı")
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="ağırlık", hue="cinsiyet", style="madalya", data=veri) # 4 boyutlu görselleştirme
plt.xlabel("Sporcuların Boyu")
plt.ylabel("Sporcuların Ağırlığı")
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="ağırlık", hue="cinsiyet", style="madalya", size="yaş", sizes=(15,200), data=veri) # 5 boyutlu görselleştirme
plt.xlabel("Sporcuların Boyu")
plt.ylabel("Sporcuların Ağırlığı")
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

# Saçılım Grafiği - Doğrusal Regresyon

```python
sns.regplot(x="boy", y="ağırlık", data=veri, marker='+', scatter_kws={'alpha':0.2})
plt.title("Boy ve ağırlık dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="ağırlık", hue="madalya", data=veri, palette="Set1")
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

```python
sns.scatterplot(x="boy", y="ağırlık", hue="madalya", data=veri, palette="rocket")
plt.title("Madalyaya Göre Boy-Ağırlık Dağılımı")
plt.show()
```

# ÇİZGİ GRAFİĞİ

```python
sns.lineplot(x='boy', y='ağırlık', data=veri)
plt.title("Boy ve Ağırlık")
plt.show()
```

# Çizgi Grafiği - Kategorik

```python
sns.lineplot(x='boy', y='ağırlık', hue="cinsiyet", data=veri)
plt.title("Cinsiyete göre boy ve ağırlık ilişkileri")
plt.show()
```

```python
sns.lineplot(x='boy', y='ağırlık', hue="madalya", data=veri)
plt.title("Madalya tipine göre boy ve ağırlık ilişkileri")
plt.show()
```

```python
sns.lineplot(x='boy', y='ağırlık', hue="sezon", data=veri)
plt.title("Sezona göre boy ve ağırlık ilişkileri")
plt.show()
```

# HİSTOGRAM

```python
sns.displot(veri, x="ağırlık")
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

# Histogram - Kategorik

```python
sns.displot(veri, x="ağırlık", hue="cinsiyet")
plt.ylabel("Frekans")
plt.title("Cinsiyete göre ağırlık histogramı")
plt.show()
```

```python
sns.displot(veri, x="ağırlık", col="cinsiyet", multiple="dodge")
plt.show()
```

# İki Boyutlu Histogram

```python
sns.displot(veri, x="boy", y="ağırlık", kind="kde") # son parametre tür parametresi istatistiksel bir ifadedir
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.title("Boy-Ağırlık Histogramı")
plt.show()
```

```python
sns.displot(veri, x="boy", y="ağırlık", hue="cinsiyet", kind="kde") 
plt.xlabel("Boy")
plt.ylabel("Ağırlık")
plt.title("Cinsiyete Göre Ayrıştırılmış Boy-Ağırlık Histogramı")
plt.show()
```

# ÇUBUK GRAFİĞİ

```python
sns.barplot(x="madalya", y="boy", data=veri)
plt.title("Madalyalara Göre Boy Grafiği")
plt.show()
```

```python
sns.barplot(x="madalya", y="boy", hue="cinsiyet", data=veri)
plt.title("Madalyalara Göre Boy Grafiği")
plt.show()
```

```python
sns.catplot(x="madalya", y="yaş", hue="cinsiyet", col="sezon", data=veri, kind="bar", height=4, aspect=0.7) # Kategorize etme
plt.show()
```

# KUTU GRAFİĞİ

```python
sns.boxplot(x="sezon", y="boy", data=veri)
plt.show()
```

```python
sns.boxplot(x="sezon", y="boy", hue="cinsiyet", palette="Set2", data=veri)
plt.show()
```

```python
veri_gecici = veri.loc[:,["boy", "ağırlık", "yaş"]]
sns.boxplot(data=veri_gecici, orient="h", palette="Set1") # Orient kutu grafiğinin yönünü belirler, h(horizantal)->yatay
plt.show()
```
